```python
import os
import numpy as np
import pandas as pd
import pprint
```

```python
# 設定 data_path
dir_data = './data/'
```

```python
f_app = os.path.join(dir_data, 'ExchangeRate@202002261600.csv')
print('Path of read in data: %s' % (f_app))
app_train = pd.read_csv(f_app)
pprint.pprint(app_train)
```

### 如果沒有想法，可以先嘗試找出剛剛例子中提到的問題的答案
#### 資料的 row 數以及 column 數

```python
pprint.pprint(app_train.shape)
#column數
pprint.pprint("column數:" +str(len(app_train.columns)))
#row數
pprint.pprint("row數:" + str(len(app_train.index)))
```

#### 列出所有欄位

```python
pprint.pprint(app_train.columns)
```

#### 截取部分資料

```python
#取出相關資料
pprint.pprint(app_train.iloc[0,1])
pprint.pprint("---------------------------------")
#取出美金部份
pprint.pprint(app_train.loc["USD"])
pprint.pprint("---------------------------------")
#只取出即期欄位
pprint.pprint(app_train["即期"])
pprint.pprint("---------------------------------")
#取出前三筆資料
pprint.pprint(app_train.head(3))
pprint.pprint("---------------------------------")
#取出後三筆資料
pprint.pprint(app_train.tail(3))
```
