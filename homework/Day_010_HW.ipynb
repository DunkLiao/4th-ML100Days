```python
# 做完特徵工程前的所有準備 (與前範例相同)
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression

data_path = 'data/'
df_train = pd.read_csv(data_path + 'house_train.csv.gz')

train_Y = np.log1p(df_train['SalePrice'])
df = df_train.drop(['Id', 'SalePrice'] , axis=1)
df.head()
```

```python
#只取 int64, float64 兩種數值型欄位, 存於 num_features 中
num_features = []
for dtype, feature in zip(df.dtypes, df.columns):
    if dtype == 'float64' or dtype == 'int64':
        num_features.append(feature)
print(f'{len(num_features)} Numeric Features : {num_features}\n')
```

```python
# 削減文字型欄位, 只剩數值型欄位
df = df[num_features]
df = df.fillna(-1)
MMEncoder = MinMaxScaler()
train_num = train_Y.shape[0]
df.head()
```

# 作業2
* 續前題, 去除離群值有兩類方式 :  捨棄離群值(刪除離群的資料) 以及調整離群值,  
請試著用同樣的上下限, 改為 '捨棄離群值' 的方法, 看看結果會變好還是變差? 並試著解釋原因。

### ANS:

```python
# 做完特徵工程前的所有準備 (與前範例相同)
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression

data_path = 'data/'
df_train = pd.read_csv(data_path + 'house_train.csv.gz')

train_Y = np.log1p(df_train['SalePrice'])
df = df_train.drop(['Id', 'SalePrice'] , axis=1)

#只取 int64, float64 兩種數值型欄位, 存於 num_features 中
num_features = []
for dtype, feature in zip(df.dtypes, df.columns):
    if dtype == 'float64' or dtype == 'int64':
        num_features.append(feature)
        
# 削減文字型欄位, 只剩數值型欄位
df = df[num_features]
df = df.fillna(-1)
MMEncoder = MinMaxScaler()
train_num = train_Y.shape[0]
df.head()        
```

```python
# 將 1stFlrSF 限制在你覺得適合的範圍內, 捨棄離群值
col_name = '1stFlrSF'
pprint.pprint(df[col_name])
pprint.pprint('----------------------------------')
# 顯示統計資訊
pprint.pprint(df[col_name].describe())
pprint.pprint('----------------------------------')
pprint.pprint(df[col_name].isna().any())
pprint.pprint('----------------------------------')



#上限:平均值+2*標準差，下限:平均值-2*標準差
upper_bound = df[col_name].mean() + 2 * df[col_name].std()
lower_bound = df[col_name].mean() - 2 * df[col_name].std()
pprint.pprint('upper_bound={0},lower_bound={1}'.format(upper_bound,lower_bound))
pprint.pprint('----------------------------------')


#捨棄離群值
keep_indexs = (df[col_name]> lower_bound) & (df[col_name]< upper_bound)
df = df[keep_indexs]
train_Y = train_Y[keep_indexs]
sns.regplot(x = df[col_name], y=train_Y)
plt.show()

# 做線性迴歸, 觀察分數
train_X = MMEncoder.fit_transform(df)
estimator = LinearRegression()
cross_val_score(estimator, train_X, train_Y, cv=5).mean()
```

#### 在此案例中，捨棄離群值比調整離群值結果較好(分數較高)，因為離峰值相對整體資料而言差異過大
#### 尤其是在較大的數據方面
