```python
# 載入需要的套件
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns # 另一個繪圖-樣式套件

# 忽略警告訊息
%matplotlib inline
plt.style.use('ggplot')
import warnings
warnings.filterwarnings('ignore')

# 設定 data_path
dir_data = './data/'
```

```python
# 讀取檔案
f_app = os.path.join(dir_data, 'application_train.csv')
print('Path of read in data: %s' % (f_app))
app_train = pd.read_csv(f_app)
app_train.head()
```

### ANS:

```python
# 使用不同的 HOUSETYPE_MODE 類別繪製圖形, 並使用 subplot 排版
# 請使用 application_train.csv, 根據不同的 HOUSETYPE_MODE 對 AMT_CREDIT 繪製 Histogram
import pprint

unique_house_type = app_train['HOUSETYPE_MODE'].unique().astype(str)
pprint.pprint(unique_house_type)

nrows = len(unique_house_type)
ncols = nrows // 2
plt.figure(figsize=(10,30))
for i in range(len(unique_house_type)):
    plt.subplot(nrows, ncols, i+1)
    if unique_house_type[i] == 'nan':
        plt.hist(app_train[app_train['HOUSETYPE_MODE'].isna()]['AMT_CREDIT'])
    else:
        plt.hist(app_train[app_train['HOUSETYPE_MODE']==unique_house_type[i]]['AMT_CREDIT'])  
    plt.title(str(unique_house_type[i]))
    plt.ylabel('AMT_CREDIT')
plt.show()    
```
