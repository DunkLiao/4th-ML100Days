```python
# 匯入套件
import pandas as pd
import numpy as np
import pprint
```

```python
#匯入資料
url = "https://www.twse.com.tw/exchangeReport/TWT53U?response=html&selectType=ALL" 
df_original = pd.read_html(url, encoding='utf-8')[0]

# 重新整理欄位名稱
column_list = [str(x[1]) for x in df_original.columns]
df_original.columns = column_list

#顯示資料
df_original
```

```python
# 資料前處理
# 顯示欄位種類
import copy
df_source = copy.deepcopy(df_original)
print(df_source.dtypes.value_counts())
print("object")
print(df_source.select_dtypes(include=["object"]).apply(pd.Series.nunique, axis = 0))
print("int64")
print(df_source.select_dtypes(include=["int64"]).apply(pd.Series.nunique, axis = 0))
# 處理異常值
for col_name in df_source.columns:
    if df_source[df_source[col_name] == '--'][col_name].count() >0 :
        df_source.loc[df_source[col_name] == '--',col_name] = 0
        df_source[col_name] = df_source[col_name].apply(pd.to_numeric)

print(df_source.dtypes.value_counts())
```

```python
# 顯示統計值
pprint.pprint(df_source.describe())
```
