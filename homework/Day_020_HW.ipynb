```python
# 載入需要的套件
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns # 另一個繪圖-樣式套件
plt.style.use('ggplot')

# 忽略警告訊息
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
# 請用 numpy 建立一個 10 x 10, 數值分布自 -1.0 ~ 1.0 的矩陣並繪製 Heatmap
import pprint
matrix = np.linspace(-1.0,1.0,num=100).reshape(10,10)
pprint.pprint('matrix:')
pprint.pprint(matrix)
pprint.pprint('shape is {0}'.format(matrix.shape))
plt.figure(figsize=(10,10))
sns.heatmap(matrix,cmap='rainbow',vmax=1.0,vmin=-1.0)
# heatmap = 
plt.show()
```

```python
# 請用 numpy 建立一個 1000 x 3, 數值分布為 -1.0 ~ 1.0 的矩陣，並繪製 PairPlot 
# (上半部為 scatter, 對角線為 hist, 下半部為 density)
nrow = 1000
ncol = 3
matrix = np.linspace(-1.0,1.0,num= nrow *ncol ).reshape(nrow,ncol)
pprint.pprint('matrix:')
pprint.pprint(matrix)
pprint.pprint('shape is {0}'.format(matrix.shape))
# 隨機給予 0, 1, 2 三種標籤
indice = np.random.choice([0,1,2], size=nrow)
plot_data = pd.DataFrame(matrix, indice).reset_index()

# 繪製 seborn 進階 Heatmap
grid = sns.PairGrid(data = plot_data, size = 3, diag_sharey=False, 
                    hue = 'index', vars = [x for x in list(plot_data.columns) if x != 'index'])

# 上半部為 scatter
grid.map_upper(plt.scatter, alpha = 0.2)
# 對角線畫 histogram
grid.map_diag(plt.hist)
# 下半部放 density plot
grid.map_lower(sns.kdeplot , cmap = plt.cm.OrRd_r)

plt.show()
```

```python
# 請用 numpy 建立一個 1000 x 3, 數值分布為常態分佈的矩陣，並繪製 PairPlot 
# (上半部為 scatter, 對角線為 hist, 下半部為 density)
# 參考網址
# https://docs.scipy.org/doc/numpy-1.15.0/reference/generated/numpy.random.normal.html
nrow = 1000
ncol = 3

mu, sigma = 0, 0.1 # mean and standard deviation
rng = np.random.RandomState(42)
matrix = rng.normal(mu, sigma, nrow *ncol).reshape(nrow,ncol)

pprint.pprint('matrix:')
pprint.pprint(matrix)
pprint.pprint('shape is {0}'.format(matrix.shape))

# 隨機給予 0, 1, 2 三種標籤
indice = np.random.choice([0,1,2], size=nrow)
plot_data = pd.DataFrame(matrix, indice).reset_index()

# 繪製 seborn 進階 Heatmap
grid = sns.PairGrid(data = plot_data, size = 3, diag_sharey=False,
                    hue = 'index', vars = [x for x in list(plot_data.columns) if x != 'index'])

# 上半部為 scatter
grid.map_upper(plt.scatter, alpha = 0.2)
# 對角線畫 histogram
grid.map_diag(plt.hist)
# 下半部放 density plot
grid.map_lower(sns.kdeplot , cmap = plt.cm.OrRd_r)

plt.show()
```
